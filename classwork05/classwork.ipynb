```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
x = np.linspace(1, 11, 11)

k = 10
epsilon = np.random.normal(0, 1e-1 * k, size=x.shape)
F = k * x + epsilon

x_err = 0.5
y_err = k * x_err


k_approx = (st.mean(x*F) - st.mean(x)*st.mean(F))/(st.mean(x*x)-(st.mean(x))**2)
b_approx = st.mean(F) - k_approx * st.mean(x)

x_expand = np.array([0, 12])

plt.plot(x_expand, k_approx * x_expand + b_approx, linestyle="--", label="Approx")
plt.errorbar(x, F, xerr=x_err, yerr=y_err, linestyle="none")

plt.scatter(x, F, s=10, marker="^", label="DATA", color="black")

plt.xlabel("x, cm")
plt.ylabel(r"$\mathcal{F}, N$", rotation=0)

plt.xlim(0, 12)
plt.ylim(0, 120)

plt.grid()
plt.minorticks_on()
plt.grid(which="minor", linestyle=":")

plt.legend()

plt.title(r"Measurement of spring, $\mathcal{F} = k \cdot x$")
plt.tight_layout()
plt.savefig("output.pdf")

plt.show()
```
